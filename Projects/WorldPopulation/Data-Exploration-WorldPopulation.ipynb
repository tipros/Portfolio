```python
# Use indexing to only columns needed. ::-1 inverts the order. Column names can also be listed instead.
df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2
```

```python
# tranpose columns
df3 = df2.transpose()
df3
```
