```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Temp\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
# generate plot
df.plot(kind = 'line', title='Ice Cream Ratings', xlabel='Daily Readings', ylabel='Scores')
```

```python
# generate subplots
df.plot(kind = 'line', subplots=True, title='Ice Cream Ratings', xlabel='Daily Readings', ylabel='Scores')
```

```python
# bar chart
df.plot(kind='bar')
```

```python
# bar chart stacked
df.plot(kind='bar', stacked = True)
```

```python
# bar chart
df['Flavor Rating'].plot(kind='bar', stacked = True)
```

```python
# horizontal bar chart
df.plot.barh(stacked=True)
```

```python
# scatter plot
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='Yellow')
```

```python
# historgrams
df.plot.hist()
```

```python
# historgrams
df.plot.hist(bins=100)
```

```python
# box plot
df.boxplot()
```

```python
# area plot
df.plot.area()
```

```python
# pie chart
df.plot.pie(y='Flavor Rating', figsize=(10,6))
```
